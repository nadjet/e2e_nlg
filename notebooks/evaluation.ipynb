In this notebook, we evaluate the output of the seq2seq predictions obtained from [this notebook](seq2seq_tf.ipynb) against the e2e NLG competition [baseline](http://www.macs.hw.ac.uk/InteractionLab/E2E/#baseline) using the competition's [metrics scripts](https://github.com/tuetschek/e2e-metrics).

The evaluation scripts take as input the reference texts file and the output texts file. The [reference texts file](https://github.com/tuetschek/e2e-metrics/blob/master/example-inputs/devel-conc.txt) consists of blocks of line-separated target sentences for the same input. Each block is separated by a blank line. The [output texts file](https://github.com/tuetschek/e2e-metrics/blob/master/example-inputs/baseline-output.txt) consists of one output text per target on a separate line and in the same order as the reference texts.

So we must first convert the output and reference texts to the metrics scripts format. That includes also relexicalizing the output.

We call a pipeline of scripts that 1) reformats the input meaning representations, 2) relexicalizes the inputs and outputs, 4) prints the reference and target texts to files following the evaluation format (target texts per block), and 5) writes the output texts in the same order as the target texts.

````python
ref_file = os.path.join(out_path,"devset.csv")
out_path = "output"
````

````python
from e2e_nlg.formatting.pipeline import DataFormattingPipeline
pipeline = DataFormattingPipeline(output_file=output_file,ref_file=ref_file, output_folder=out_path)
pipeline.prepare_data()
````

This pipeline produces 3 output files: the reference inputs, the reference outputs and the output texts. The latter two are in the format described above ready for evaluation with metrics script, which can be called as follows:

 ```./measure_scores.py ref_e2e_nlg.txt pred_e2e_nlg.txt```
